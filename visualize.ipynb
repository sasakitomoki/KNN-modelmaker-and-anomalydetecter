```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%cd /content/drive/MyDrive/sport/K-NN
```

```python
video_name1 = "front_1-5"
video_name2 = "front_1-6"
position_data1 = pd.read_csv(f"dataset/{video_name1}/position_data.csv")
#threeD_position_data = pd.read_csv(f"dataset/{video_name}/3d_position_data.csv")
position_data2 = pd.read_csv(f"dataset/{video_name2}/position_data.csv")
```

##すべての表の可視化

```python
#出力したい列名を入力,すべて出力したい場合は空でよい
#column_li = ["LEFT_ANKLE_x", "RIGHT_ANKLE_x", "LEFT_HEEL_x", "RIGHT_HEEL_x","LEFT_ANKLE_y", "RIGHT_ANKLE_y", "LEFT_HEEL_y", "RIGHT_HEEL_y","LEFT_FOOT_INDEX_x", "RIGHT_FOOT_INDEX_x","LEFT_FOOT_INDEX_y", "RIGHT_FOOT_INDEX_y","LEFT_KNEE_x", "RIGHT_KNEE_x","LEFT_KNEE_y", "RIGHT_KNEE_y","LEFT_WRIST_x","RIGHT_WRIST_x","LEFT_PINKY_x", "RIGHT_PINKY_x", "LEFT_INDEX_x", "RIGHT_INDEX_x", "LEFT_THUMB_x", "RIGHT_THUMB_x","LEFT_WRIST_y","RIGHT_WRIST_y","LEFT_PINKY_y", "RIGHT_PINKY_y", "LEFT_INDEX_y", "RIGHT_INDEX_y", "LEFT_THUMB_y", "RIGHT_THUMB_y","LEFT_SHOULDER_x","RIGHT_SHOULDER_x","LEFT_SHOULDER_y","RIGHT_SHOULDER_y","LEFT_ELBOW_x","RIGHT_ELBOW_x","LEFT_ELBOW_y","RIGHT_ELBOW_y"]
column_li = []
```

```python
fig = plt.figure()
ax=[]
if len(column_li)==0:
  column_li = position_data1.columns[1:].values
  filtered_column_li = [col for col in column_li if not col.endswith("z")]#zの列名削除
  column_li = filtered_column_li

ax_num = len(column_li) + 1

for i in range(2*(ax_num-1)):
    ax.append(fig.add_subplot(ax_num , 2 ,i+1))

for i,column in enumerate(column_li):
  ax[2*i].plot(position_data1["frame"],position_data1[column])
  ax[2*i].set_title(column)
  ax[2*i+1].plot(position_data2["frame"],position_data2[column])
  ax[2*i+1].set_title(column)

fig.set_size_inches(16, ax_num * 4)
plt.show()
```
